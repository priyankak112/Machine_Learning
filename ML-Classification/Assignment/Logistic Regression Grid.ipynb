```python
import pandas as pd
```

```python
dataset=pd.read_csv('CKD.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
indep
```

```python
dep=dataset[['classification_yes']]
dep
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train[0]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid={'penalty': ['l2'],
           'solver': ['liblinear', 'saga','lbfgs','newton-cg','sag']
          }
grid=GridSearchCV(LogisticRegression(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
print("the best parameters:",grid.best_params_)
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import f1_score
f1_wght_average=f1_score(y_test,grid_pred,average='weighted')
print("The F1 score final:",f1_wght_average)
```

```python
from sklearn.metrics  import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,grid_pred)
print(report)
```

```python
from sklearn.metrics import roc_auc_score
scr=roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
print("The roc score:",scr)
```
