```python
#pandas is python library used to work with structured data like tables,database,spreadsheets
import pandas as pd
#Reading the input csv file and assign it to dataset
import matplotlib.pyplot as plt
df=pd.read_csv("Salary_Data.csv")
df
```

```python
plt.scatter(df['YearsExperience'], df['Salary'], color='orange')
plt.title('YearsExperience vs Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary ($)')
plt.show()
```

```python
#assigning YearsExperience data to input variable/independent variable
independent=df[["YearsExperience"]]
independent
```

```python
#assigning Salary data to dependent variable
dependent=df[["Salary"]]
dependent
```

```python
#sklearn=python library,model_selection=module,train_test_split=function to split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#MODEL CREATION OR LEARNING PHASE
from sklearn.linear_model import LinearRegression #sklearn-python library,linear_model-module,linearregression-class
regressor=LinearRegression() #regressor-object for the class LinearRegression
regressor.fit(x_train,y_train) #training the model to learn the relationship between these 2 variables
```

```python
#now model created and  checking weight 
weight=regressor.coef_
weight
```

```python
#printing bias value(origin value/intercept)
bias=regressor.intercept_
bias
```

```python
#use the trained model to predict the output for test data
y_predict=regressor.predict(x_test)
```

```python
#find r^2 using r2_score function  to check how well trained model predicted value matches actual value 
from sklearn.metrics import r2_score
r_Score=r2_score(y_test,y_predict)
r_Score

#if r_Score near to 1 then model is good
#if r_Score near to 0 then model is weak
```

```python
#no save the best model
import pickle
filename="finalized_model_linear.sav"
#serialization using pickle where pyhton model convert into binary and write in the file
pickle.dump(regressor,open(filename,"wb"))
```

```python
#Test model- using real values-first load the model using pickle and  deserialize it - read the binary format to  python model then use predict function to find the result
loaded_model=pickle.load(open(filename,'rb'))
result=loaded_model.predict([[15]])
result
```
