```python
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset['charges']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['squared_error','absolute_error','friedman_mse'],
            'max_features':['sqrt','log2'],
            'n_estimators':[10,100]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1,cv=3)
grid.fit(indep,dep)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
re=grid.cv_results_
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
fin_pred=grid.predict([[19,0,27.900,0,1]])
print(f"radom Forest gridsearch prediction is:{fin_pred}")
```
